```python
import tensorflow as tf
import keras
from keras.layers import Input, Conv2D, MaxPool2D, Add, BatchNormalization, Activation, Dense
```

## The Model

```python
img_size = (224, 224, 3)

X_input = Input(shape=img_size)
X = Conv2D(filters = 64, kernel_size = (7,7), strides = (2,2))(X_input)
X = BatchNormalization()(X)
X = Activation('relu')(X)

X = MaxPool2D(pool_size = (2,2))(X)

X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 64, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 128, kernel_size = (3,3), strides = (2,2))(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X = Activation('relu')(X2)

X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 128, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), strides = (2,2))(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 256, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 512, kernel_size = (3,3), strides = (2,2))(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 512, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X = Activation('relu')(X2)

X2 = Conv2D(filters = 512, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 512, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X2 = Conv2D(filters = 512, kernel_size = (3,3), padding = "same")(X)
X2 = BatchNormalization()(X2)
X2 = Activation('relu')(X2)
X2 = Conv2D(filters = 512, kernel_size = (3,3), padding = "same")(X2)
X2 = BatchNormalization()(X2)
X2 = Add()([X2, X])
X = Activation('relu')(X2)

X = tf.keras.layers.GlobalAveragePooling2D()(X)
X = Dense(units = 1000, activation = 'softmax')(X)

resnet = tf.keras.Model(inputs = X_input, outputs = X)

resnet.summary()
```

## Data Preparation

**Resnet uses random cropping and random horizontal flip**

```python
import numpy as np

def random_crop(img, random_crop_size):
    assert img.shape[2] == 3
    height, width = img.shape[0], img.shape[1]
    dy, dx = random_crop_size
    x = np.random.randint(0, width - dx + 1)
    y = np.random.randint(0, height - dy + 1)
    return img[y:(y+dy), x:(x+dx), :]

def crop_generator(batches, crop_length):
    while True:
        batch_x, batch_y = next(batches)
        batch_crops = np.zeros((batch_x.shape[0], crop_length, crop_length, 3));
        for i in range(batch_x.shape[0]):
            batch_crops = random_crop(batch_x[i], (crop_length, crop_length))
        yield(batch_crops, batch_y)
```

```python
from keras.preprocessing.image import ImageDataGenerator

TRAIN_DIR = "../input/imagenetmini-1000/imagenet-mini/train/"
VAL_DIR = "../input/imagenetmini-1000/imagenet-mini/val/"
batch_size = 64

train_generator = ImageDataGenerator(horizontal_flip=True)
val_generator = ImageDataGenerator(horizontal_flip=True)

train_gen = train_generator.flow_from_directory(TRAIN_DIR, 
                                               target_size=(224, 224), 
                                               batch_size=batch_size, 
                                               class_mode='categorical')
val_gen = val_generator.flow_from_directory(VAL_DIR, 
                                               target_size=(224, 224), 
                                               batch_size=batch_size, 
                                               class_mode='categorical')
```

## Training the model

```python
resnet.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
from livelossplot import PlotLossesKeras
history = resnet.fit_generator(train_gen,
                    epochs = 3,
                    steps_per_epoch = 34745 // batch_size,
                    validation_data = val_gen,
                    validation_steps = 3923 // batch_size,
                    callbacks = [PlotLossesKeras()],
                    verbose = 1)
```
